## 1. Problem Statement

This project analyzes banking customer accounts, transactions, and loan data to understand financial behavior, evaluate loan performance, and identify default risks. The goal is to clean and explore the datasets, generate descriptive statistics, and visualize trends to support data-driven decisions. By examining balances, transaction patterns, and loan outcomes, the project provides insights that help improve customer segmentation, financial monitoring, and reduce loan default risks.

**Key Focus Areas**

1. Analyze customer accounts to study balance distribution and account activity

2. Explore transaction patterns to identify spending behavior and trend changes

3. Evaluate loan performance and detect factors associated with default risks

4. Visualize financial trends to support data-driven banking decisions

5. Provide actionable insights for customer segmentation and financial risk control
